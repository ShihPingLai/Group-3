```python
#1 [transforming the text files of mixture into list variables]

length = 720832  #要想辦法弄出來

f1 = open("sound_ext1.txt", "r")
f2 = open("sound_ext2.txt", "r")
f3 = open('sound1.txt', 'r') 
f4 = open('sound2.txt', 'r')
f5 = open("time.txt", "r")

sound1 = []
sound2 = []
ext1 = []
ext2 = []
t = []

print('  Reading the txt. files and creating the list variables...')

for i in range(length):
    ext1.append(float(f1.readline()))
    ext2.append(float(f2.readline()))
    sound1.append(float(f3.readline()))
    sound2.append(float(f4.readline()))
    t.append(float(f5.readline()))
    
print('  Done!')

f1.close()
f2.close()
f3.close()
f4.close()
f5.close()
```

```python
#2 [Comparing the extracted sound with the original one]

import matplotlib.pyplot as plt

plt.plot(t, ext1)
plt.title("Channel_1_extracted")
plt.show()

plt.plot(t, ext2)
plt.title("Channel_2_extracted")
plt.show()

# sound signals imported from original file
plt.plot(t, sound1)
plt.title("Channel_1_original")
plt.show()

plt.plot(t, sound2)
plt.title("Channel_2_original")
plt.show()
```

```python
#3 [python numerical error(32 bits at most)]

import numpy as np

#這裡的ext、sound都是list，list不能直接減，所以要再轉一下array（原本有跟混沌的H計算到就自動變成array了，所以才可以減，但這裡沒有）
ext_array1 = np.array(ext1)
ext_array2 = np.array(ext2)


error1 = ext_array1 - sound1
plt.plot(t, error1)
plt.title("error_1")
plt.show()

error2 = ext_array2 - sound2 
plt.plot(t, error2)
plt.title("error_2")
plt.show()


#混沌變大了，error就變大了
```
